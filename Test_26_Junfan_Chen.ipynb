```python
def DeBruijnGraph(text, k):
    prefix_dictionary = {}
    n = len(text)
    
    for i in range(n - k + 1):
        if text[i : i+k-1] not in prefix_dictionary.keys():
            prefix_dictionary[text[i : i+k-1]] = ' -> ' + text[i+1 : i+k]
        else:
            prefix_dictionary[text[i : i+k-1]] += ',' + text[i+1 : i+k]
    return prefix_dictionary
```

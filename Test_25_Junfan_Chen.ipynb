```python
def readIn(nameOftxt):    
    with open(nameOftxt, 'r') as f:
        list_pattern = f.readlines()
        n = len(list_pattern)
        for i in range(n):
            if list_pattern[i][-1] == '\n':
                list_pattern[i] = list_pattern[i][:-1]
    return list_pattern
```

```python
def AdjacentGraph(list_pattern):
    n = len(list_pattern)
    k = len(list_pattern[0])
    overlap_pattern = []
    
    for i in range(n):
        assert k == len(list_pattern[i])
        surffix = list_pattern[i][1:]
        for j in range(n):
            assert k == len(list_pattern[j])
            prefix = list_pattern[j][:k-1]
            if surffix == prefix:
                overlap_pattern.append((list_pattern[i] + ' -> ' + list_pattern[j]))
    overlap_pattern.sort()
    return overlap_pattern
```

```python
if __name__ == '__main__':
    pattern = readIn('rtext.txt')
    overlap_pattern = AdjacentGraph(pattern)
    n = len(overlap_pattern)
    for i in range(n):
        print(overlap_pattern[i])
```

```python
def kmers_to_gene(kmers,k):
    matches={}
    #make a dictionary for when prefix matches suffix
    for i in kmers:
        for z in kmers:
            if i[0:k-1]==z[1:k]:
                matches[i]=z
    #find the starting node which is the node with a prefic that doesn't match any suffix
    start=[]
    for i in kmers:
        if i not in matches:
            start.append(i)
    
    gene=str(start[0])
    # match first k-1 characters of elemint in kmers to the last (k-1) characters
    while len(gene)<len(kmers)+k-1:
        for i in kmers:
            if i[0:k-1]==gene[(1-k):]:
                gene+=i[-1]
    return gene
    
```

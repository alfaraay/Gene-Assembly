```python
def readIn(fileName):
    with open(fileName, 'r') as f:
        composition = f.readlines()
    n = len(composition)
    for i in range(n):
        if composition[i][-1] == '\n':
            composition[i] = composition[i][:-1]
    composition.sort()
    return composition
```

```python
def DeBruijnGraph(composition):
    adjacent_dictionary = {}
    n = len(composition)
    k = len(composition[0])
    
    for i in range(n):
        assert k == len(composition[i])
        if composition[i][:-1] not in adjacent_dictionary.keys():
            adjacent_dictionary[composition[i][:-1]] = ' -> ' + composition[i][1:]
        else:
            adjacent_dictionary[composition[i][:-1]] += ',' + composition[i][1:]
    return adjacent_dictionary
```

```python
if __name__ == '__main__':
    composition = readIn('rtext.txt')
    DeBruijn = DeBruijnGraph(composition)
    for item in DeBruijn:
        print(item + DeBruijn[item])
```

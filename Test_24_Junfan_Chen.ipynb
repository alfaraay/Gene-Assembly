```python
def conctenate_string(PatternList):
    k = len(PatternList[0])
    n = len(PatternList)
    text = PatternList[0]
    for i in range(n - 1):
        text = text + PatternList[i + 1][-1]
    return text
```

```python
def readIn(nameOftxt):    
    with open(nameOftxt, 'r') as f:
        list_pattern = f.readlines()
        n = len(list_pattern)
        for i in range(n):
            if list_pattern[i][-1] == '\n':
                list_pattern[i] = list_pattern[i][:-1]
    return list_pattern
```
